```python
# Exclude the "normal" event type
df_filtered = df.drop(columns=['normal', 'total'])

# Plot the filtered data
df_filtered.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribution of Non-Normal Network Events by File')
plt.xlabel('File')
plt.ylabel('Count')
plt.show()
```

```python
# Analyze the distribution of each event type across files
df_filtered.plot(kind='bar',stacked=True, figsize=(10, 6))
plt.title('Distribution of Each Event Type by File')
plt.xlabel('File')
plt.ylabel('Count')
plt.show()

```

```python
from sklearn.ensemble import IsolationForest

# Fit the Isolation Forest model to the DataFrame
model = IsolationForest(contamination=0.1)  # Adjust the contamination parameter as needed
df_filtered['anomaly_score'] = model.fit_predict(df_filtered)

# Plot the anomaly scores
plt.figure(figsize=(10, 6))
plt.bar(df_filtered.index, df_filtered['anomaly_score'])
plt.title('Anomaly Scores by File')
plt.xlabel('File')
plt.ylabel('Anomaly Score')
plt.show()
```

```python
# Filter the anomalous files
anomalies = df_filtered[df_filtered['anomaly_score'] == -1]

# Display the contents of the anomalous files
print(anomalies)
```

```python
normal_files = df_filtered[df_filtered['anomaly_score'] == 1]

# Plot comparison
anomalies.mean().plot(kind='bar', color='red', label='Anomalies', alpha=0.5)
normal_files.mean().plot(kind='bar', color='blue', label='Normal', alpha=0.5)
plt.title('Comparison of Event Distribution: Anomalous vs Normal Files')
plt.xlabel('Event Type')
plt.ylabel('Average Count')
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import LocalOutlierFactor

# Applying LOF
lof_model = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
df_filtered['lof_score'] = lof_model.fit_predict(df_filtered.drop(columns=['anomaly_score']))
```

```python
df_filtered['exploit_to_total_ratio'] = df_filtered['exploits'] / df_filtered.sum(axis=1)
df_filtered['dos_to_total_ratio'] = df_filtered['dos'] / df_filtered.sum(axis=1)
```

```python
# Create a comparison column to see where both models agree on anomalies
df_filtered['both_anomalies'] = (df_filtered['anomaly_score'] == -1) & (df_filtered['lof_score'] == -1)

# Check the files where both models agree
both_anomalies = df_filtered[df_filtered['both_anomalies'] == True]
print("Files flagged as anomalies by both models:\n", both_anomalies)
```

```python
# Count the number of anomalies detected by each model
anomaly_counts = {
    'Isolation Forest Anomalies': (df_filtered['anomaly_score'] == -1).sum(),
    'LOF Anomalies': (df_filtered['lof_score'] == -1).sum(),
    'Both Models Anomalies': both_anomalies.shape[0]
}

print("Anomaly Counts:\n", anomaly_counts)
```

```python
# Group by whether the file is an anomaly or not
grouped = df_filtered.groupby(['both_anomalies'])[['exploit_to_total_ratio', 'dos_to_total_ratio']].mean()

print("Average Ratios for Anomalous vs. Non-Anomalous Files:\n", grouped)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = 0.3 * np.random.randn(100, 2)
X_train = np.r_[X + 2, X - 2]  # Normal data
X_outliers = np.random.uniform(low=-4, high=4, size=(20, 2))  # Outliers

# Combine normal data and outliers
X_data = np.r_[X_train, X_outliers]

# Visualize the data
plt.scatter(X_data[:, 0], X_data[:, 1], color='blue')
plt.title("Data with Outliers")
plt.show()
```

```python
from scipy.spatial import distance_matrix

def knn_anomaly_detection(X, k=5, threshold=1.5):
    # Compute distance matrix
    distances = distance_matrix(X, X)

    # Sort distances and take the first k distances for each point
    sorted_distances = np.sort(distances, axis=1)[:, 1:k+1]

    # Calculate the mean distance to the k nearest neighbors
    avg_distances = np.mean(sorted_distances, axis=1)

    # Identify anomalies (points with avg distance greater than threshold)
    anomalies = avg_distances > threshold

    return anomalies, avg_distances

# Apply k-NN based anomaly detection
k = 5
threshold = 1.5  # Adjust based on your data distribution
anomalies, avg_distances = knn_anomaly_detection(X_data, k, threshold)

# Visualize anomalies
plt.scatter(X_data[:, 0], X_data[:, 1], color='blue', label='Normal')
plt.scatter(X_data[anomalies, 0], X_data[anomalies, 1], color='red', label='Anomalies')
plt.legend()
plt.title("Anomaly Detection using k-NN")
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score

# Assume true labels: 0 for normal, 1 for anomalies
y_true = np.zeros(len(X_data))
y_true[len(X_train):] = 1  # Mark the outliers

# Predicted labels from our model
y_pred = anomalies.astype(int)

# Calculate precision and recall
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
```

```python
threshold = 1.2  # Lower the threshold
anomalies, avg_distances = knn_anomaly_detection(X_data, k, threshold)

# Re-evaluate precision and recall
y_pred = anomalies.astype(int)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
```
