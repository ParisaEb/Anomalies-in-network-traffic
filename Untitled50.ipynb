```python
import pandas as pd
```

```python
data=pd.read_csv('ratings_Electronics (1).csv',names=['userID',  'ProductId','Score','Time'])
```

```python
data.columns
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set plot style for a cleaner look
sns.set_style("whitegrid")

# Create the countplot with added elements
plt.figure(figsize=(10, 6))
sns.countplot(x='Score', data=data, palette='coolwarm', edgecolor='black')

# Adding labels
plt.xlabel('Product Rating', fontsize=14, fontweight='bold')
plt.ylabel('Frequency', fontsize=14, fontweight='bold')
plt.title('Distribution of Product Ratings', fontsize=16, fontweight='bold')

# Highlight the highest frequency bar
max_value = data['Score'].value_counts().max()
for patch in plt.gca().patches:
    if patch.get_height() == max_value:
        patch.set_facecolor('red')

# Add gridlines and improve axis
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```

```python

sns.countplot(x='Score', data=data, color='skyblue', edgecolor='black')
plt.xlabel('The rating')
plt.ylabel('Frequency')
plt.title('Distribution of Product Ratings')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
top_users=data['userID'].value_counts().head(10)

print(top_users)
```

```python
x=top_users.index
y=top_users.values
plt.figure(figsize=(12, 5))
sns.barplot(x=top_users.index, y=top_users.values,palette='viridis')
plt.xlabel('The users')
plt.xticks(rotation=45)
plt.ylabel('the number of review')
```

```python
top_ten_product=data['ProductId'].value_counts().head(10)
plt.xlabel('The Product Id')
plt.xticks(rotation=45)
plt.ylabel('the rating')
sns.barplot(x=top_ten_product.index,y=top_ten_product.values)

```

```python
data['Time'] = pd.to_datetime(data['Time'], unit='s')
plt.figure(figsize=(12, 5))

data.groupby(data['Time'].dt.to_period('M')).size().plot(kind='line', linewidth=2.5, color='blue')
plt.show()
```
